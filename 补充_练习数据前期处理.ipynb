## 对缺失值、格式不对的值的处理
打开数据发现：数据格式不对，有NAN，原本应该是小数点却有逗号

```python
import pandas as pd

df = pd.read_csv("../参考/countries of the world.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.dropna() # 清除掉缺失值
df = df.iloc[:,2:].apply(lambda x: x.astype(str).str.replace(',','.').astype(float)) # 转换
```
